# Questão 1:

<p style="text-align: justify;">
O Deep Learning é uma área da Inteligência Artificial que se inspira no funcionamento do cérebro humano para aprender e reconhecer padrões complexos. Ele é utilizado em muitas tecnologias que usamos todos os dias, como assistentes de voz, aplicativos de reconhecimento de imagem e até mesmo na filtragem de spam no e-mail.
</p>

<p style="text-align: justify;">
Neste fórum, você vai refletir sobre onde o Deep Learning está presente no seu cotidiano e como ele melhora algumas atividades.
</p>

<p style="text-align: justify;">
Dê um exemplo de uma aplicação que você usa no dia a dia e que pode utilizar Deep Learning. Explique como você acha que essa tecnologia está sendo usada para melhorar a experiência ou o funcionamento dessa ferramenta.
</p>


# **Resposta:**

Grande parte do meu dia a dia passo no computador, seja desenvolvendo algo, estudando, pesquisando ou assistindo a vídeos. Nos aplicativos que utilizo, sempre há alguma aplicação de Deep Learning para melhorar a interação do usuário com a plataforma. Alguns exemplos são:

- **YouTube:** quando recomenda outros vídeos com base no meu histórico de visualizações e preferências.

- **Google:** ao pesquisar por um produto para comprar, ele exibe anúncios recomendados com base nos meus interesses e buscas anteriores.

- **Google Tradutor:** utilizo essa ferramenta para estudar inglês. O sistema usa Deep Learning para entender melhor o contexto das frases e fornecer traduções mais precisas.

- **Spotify:** diariamente, uso a tecnologia de recomendação da plataforma para conhecer novos artistas. O Spotify emprega redes neurais profundas para sugerir músicas e criar playlists personalizadas, analisando padrões de audição, comparando com perfis similares e reconhecendo características sonoras, como ritmo e melodia, melhorando assim a experiência do usuário.
