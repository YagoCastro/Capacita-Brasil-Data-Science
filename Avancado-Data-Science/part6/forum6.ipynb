# **Questo 3:**
O que voc acha que leva um modelo a funcionar muito bem nos dados de treino, mas falhar com novos dados? Que consequncias prticas isso pode ter e como podemos evitar esse problema?

# **Resposta:**

iSSO acontece quando um modelo se ajusta demais aos dados de treino, capturando rudos e padres especficos daquele conjunto de dados, mas perdendo a capacidade de generalizar para novos dados.

**Causas do por que iso pode acontecer:**

- Modelo muito complexo: Se um modelo tem muitos parmetros (como rvores muito profundas no caso do Random Forest), ele pode memorizar os dados de treino em vez de aprender padres teis.

- Poucos dados de treino: Se o conjunto de treino  pequeno, o modelo pode aprender padres especficos que no se repetem nos dados reais.

- Dados com rudo ou outliers: Se houver muitos valores atpicos, o modelo pode tentar ajust-los, prejudicando a generalizao.

**Consequncias prticas do overfitting:**

- Falha na previso: O modelo tem um desempenho ruim com novos dados, tornando-se intil para previses do mundo real.

- Decises erradas: Em aplicaes crticas (como diagnstico mdico ou manuteno preditiva), decises baseadas em um modelo sobreajustado podem levar a erros graves.

- Maior custo computacional: Modelos muito complexos podem demandar mais tempo e recursos para treinamento e inferncia sem trazer benefcios reais.

**Como evitar o overfitting?**

- Dividir bem os dados: Utilizar treino, validao e teste para avaliar corretamente o modelo antes de implant-lo.

- Regularizao:Em Random Forest, pode-se limitar a profundidade das rvores e definir um nmero mnimo de amostras por folha para evitar divises excessivas.

- Mais dados e melhor qualidade: Se possvel, coletar mais dados e remover rudos e outliers que possam estar distorcendo o aprendizado.

- Tcnicas de ensemble learning: Random Forest j  um mtodo de ensemble, mas pode-se testar bagging e boosting para melhorar a generalizao.

- Validao cruzada (Cross-Validation): Ajuda a garantir que o modelo no est apenas memorizando os dados de treino.
