# **Questão 2:**
Por que é crucial separar os dados entre treino e teste na construção de um modelo de aprendizado de máquina? O que pode acontecer se não fizermos essa separação?

# **Resposta:**

A separação entre treino e teste dos dados é extremamente importante, pois garante a avaliação do modelo com dados que ele ainda não viu. Isso permite identificar se o modelo gerou ou não overfitting, ou seja, se ele acabou decorando os dados de treino em vez de aprender, de fato, o comportamento dos dados, falhando assim em generalizar a solução encontrada.

Na aplicação de modelos de aprendizagem de máquina, é crucial garantir um equilíbrio entre viés e variância. Um modelo que se ajusta excessivamente aos dados de treino pode ter um desempenho significativamente inferior ao ser exposto a novos dados. No entanto, também é importante evitar underfitting, que ocorre quando o modelo não consegue aprender com os dados de treino, resultando em métricas insatisfatórias e comprometendo a solução do projeto.

# **Questão 3:**
O que você acha que leva um modelo a funcionar muito bem nos dados de treino, mas falhar com novos dados? Que consequências práticas isso pode ter e como podemos evitar esse problema?

# **Resposta:**

iSSO acontece quando um modelo se ajusta demais aos dados de treino, capturando ruídos e padrões específicos daquele conjunto de dados, mas perdendo a capacidade de generalizar para novos dados.

**Causas do por que iso pode acontecer:**

- Modelo muito complexo: Se um modelo tem muitos parâmetros (como árvores muito profundas no caso do Random Forest), ele pode memorizar os dados de treino em vez de aprender padrões úteis.

- Poucos dados de treino: Se o conjunto de treino é pequeno, o modelo pode aprender padrões específicos que não se repetem nos dados reais.

- Dados com ruído ou outliers: Se houver muitos valores atípicos, o modelo pode tentar ajustá-los, prejudicando a generalização.

**Consequências práticas do overfitting:**

- Falha na previsão: O modelo tem um desempenho ruim com novos dados, tornando-se inútil para previsões do mundo real.

- Decisões erradas: Em aplicações críticas (como diagnóstico médico ou manutenção preditiva), decisões baseadas em um modelo sobreajustado podem levar a erros graves.

- Maior custo computacional: Modelos muito complexos podem demandar mais tempo e recursos para treinamento e inferência sem trazer benefícios reais.

**Como evitar o overfitting?**

- Dividir bem os dados: Utilizar treino, validação e teste para avaliar corretamente o modelo antes de implantá-lo.

- Regularização:Em Random Forest, pode-se limitar a profundidade das árvores e definir um número mínimo de amostras por folha para evitar divisões excessivas.

- Mais dados e melhor qualidade: Se possível, coletar mais dados e remover ruídos e outliers que possam estar distorcendo o aprendizado.

- Técnicas de ensemble learning: Random Forest já é um método de ensemble, mas pode-se testar bagging e boosting para melhorar a generalização.

- Validação cruzada (Cross-Validation): Ajuda a garantir que o modelo não está apenas memorizando os dados de treino.
