# **Questão 1:**

A engenharia de prompts é a prática de criar perguntas e comandos que ajudem um modelo de linguagem a entender exatamente o que você quer saber. Um bom prompt deve ser claro, direto e detalhado para evitar respostas confusas ou incompletas. Neste fórum, você vai pensar em como elaborar um prompt que permita que uma IA responda com precisão e traga as informações de que você precisa.

Imagine que você está usando um assistente virtual (como Siri, Alexa ou o ChatGPT) e quer pedir ajuda para estudar para uma prova de Matemática. Se você fizer uma pergunta muito vaga, como “Me ajude com Matemática”, a IA não saberá exatamente o que você precisa, pois esse pedido é muito amplo. Em vez disso, você deve criar um prompt mais claro, como “Explique como calcular a área de um triângulo com exemplos”.
