# Questo.1.
Suponha que voc tem um array que armazena os nomes de cinco frutas: [Ma, Banana, Cereja, Damasco, Uva]. 
Como voc adicionaria uma nova fruta ao final do array e como removeria a fruta da terceira posio?

## Mtodo.1:

```python
# Adicionando com o operador de adio:
frutas_1 = frutas_1 + ['Laranja']
```

```python
# Apresentao do resultado:
print(frutas_1)
```

## Mtodo.2:

```python
# Utilizando o mtodo append:
frutas_2.append('Laranja')
```

```python
# Apresentao do resultado:
print(frutas_2)
```

## Mtodo.3:

```python
# Utilizando o mtodo extend:
frutas_3.extend(['Laranja'])
```

```python
# Apresentao do resultado:
print(frutas_3)
```

# Questo.3.
Considere uma pilha utilizada para gerenciar uma sequncia de livros. 
Se voc adicionar trs livros  pilha e depois remover o livro no topo, qual livro ser o prximo no topo da pilha? D um exemplo com os livros Livro A, Livro B e Livro C.


Fonte: https://terminaldeinformacao.com/2013/07/23/entendendo-pilha-e-fila/

Seguindo a lgica de funcionamento de uma pilha na programao, podemos imaginar uma pilha de papeis, onde o primeiro elemento que ser retirado vai ser do topo, logo se o Livro C est no topo o prximo elemento  o Livro B.

```python
pilha_livros = []
```

```python
# Adicionando os Livros na pilha
# Como se estivesse deixando os livros em cima de uma mesa

pilha_livros.append('LivroA')
pilha_livros.append('LivroB')
pilha_livros.append('LivroC')

# Apresentao da pilha
print(pilha_livros)
```

```python
# Removendo o timo elemento do array que  o primeiro elemento que vai sair da pilha
pilha_livros.pop()

# Apresentao da pilha aps a primeira remoo
print(pilha_livros)
```

```python
# Removendo o prximo elemento que ser o 'LivroB'
pilha_livros.pop()

# Apresentao da pilha aps a segunda remoo
print(pilha_livros)
```

# Questo.4.
Imagine uma fila utilizada para gerenciar a ordem de atendimento em um consultrio mdico. 
Se quatro pacientes so adicionados  fila e dois so atendidos, quem sero os prximos dois pacientes na ordem de atendimento? Considere os pacientes Paciente 1, Paciente 2, Paciente 3 e Paciente 4.


Fonte: https://terminaldeinformacao.com/2013/07/23/entendendo-pilha-e-fila/

Considerando a lgica de funcionamento da fila, se os pacientes 1 e 2 j foram atendidos, pois foram os dois primeiros adicionados  fila, os prximos a serem atendidos sero os pacientes 3 e 4, respectivamente.

```python
fila_atendimento = []
```

```python
# Adicionando os Pacientes:
fila_atendimento.append('Paciente_1')
fila_atendimento.append('Paciente_2')
fila_atendimento.append('Paciente_3')
fila_atendimento.append('Paciente_4')
```

```python
# Removendo o Paciente 1, pois ele foi o primeiro a entrar na fila:
fila_atendimento.pop(0)

# Apresentaao:
print(fila_atendimento)
```

```python
# Removendo o Paciente 2, como o Paciente 1 j saiu da fila o Paciente 2 est na posio 0:
fila_atendimento.pop(0)

# Apresentaao:
print(fila_atendimento)
```
