# Questão.1.
Suponha que você tem um array que armazena os nomes de cinco frutas: [“Maçã”, “Banana”, “Cereja”, “Damasco”, “Uva”]. 
Como você adicionaria uma nova fruta ao final do array e como removeria a fruta da terceira posição?

```python
# Criando os arrays:
frutas_1 = ['Maçã', 'Banana', 'Cereja', 'Damasco', 'Uva']
frutas_2 = ['Maçã', 'Banana', 'Cereja', 'Damasco', 'Uva']
frutas_3 = ['Maçã', 'Banana', 'Cereja', 'Damasco', 'Uva']
frutas_4 = ['Maçã', 'Banana', 'Cereja', 'Damasco', 'Uva']
```

## Método.1:

```python
# Adicionando com o operador de adição:
frutas_1 = frutas_1 + ['Laranja']
```

```python
# Apresentação do resultado:
print(frutas_1)
```

## Método.2:

```python
# Utilizando o método append:
frutas_2.append('Laranja')
```

```python
# Apresentação do resultado:
print(frutas_2)
```

## Método.3:

```python
# Utilizando o método extend:
frutas_3.extend(['Laranja'])
```

```python
# Apresentação do resultado:
print(frutas_3)
```

## Método.4:

```python
# Utilizando o método insert junto com o método len:
frutas_4.insert(len(frutas_4) + 1, 'Laranja')
```

```python
# Apresentação do resultado:
print(frutas_4)
```

# Questão.3.
Considere uma pilha utilizada para gerenciar uma sequência de livros. 
Se você adicionar três livros à pilha e depois remover o livro no topo, qual livro será o próximo no topo da pilha? Dê um exemplo com os livros “Livro A”, “Livro B” e “Livro C”.


Fonte: https://terminaldeinformacao.com/2013/07/23/entendendo-pilha-e-fila/

Seguindo a lógica de funcionamento de uma pilha na programação, podemos imaginar uma pilha de papeis, onde o primeiro elemento que será retirado vai ser do topo, logo se o Livro C está no topo o próximo elemento é o Livro B.

```python
pilha_livros = []
```

```python
# Adicionando os Livros na pilha
# Como se estivesse deixando os livros em cima de uma mesa

pilha_livros.append('LivroA')
pilha_livros.append('LivroB')
pilha_livros.append('LivroC')

# Apresentação da pilha
print(pilha_livros)
```

```python
# Removendo o útimo elemento do array que é o primeiro elemento que vai sair da pilha
pilha_livros.pop()

# Apresentação da pilha após a primeira remoção
print(pilha_livros)
```

```python
# Removendo o próximo elemento que será o 'LivroB'
pilha_livros.pop()

# Apresentação da pilha após a segunda remoção
print(pilha_livros)
```

# Questão.4.
Imagine uma fila utilizada para gerenciar a ordem de atendimento em um consultório médico. 
Se quatro pacientes são adicionados à fila e dois são atendidos, quem serão os próximos dois pacientes na ordem de atendimento? Considere os pacientes “Paciente 1”, “Paciente 2”, “Paciente 3” e “Paciente 4”.


Fonte: https://terminaldeinformacao.com/2013/07/23/entendendo-pilha-e-fila/

Considerando a lógica de funcionamento da fila, se os pacientes 1 e 2 já foram atendidos, pois foram os dois primeiros adicionados à fila, os próximos a serem atendidos serão os pacientes 3 e 4, respectivamente.

```python
fila_atendimento = []
```

```python
# Adicionando os Pacientes:
fila_atendimento.append('Paciente_1')
fila_atendimento.append('Paciente_2')
fila_atendimento.append('Paciente_3')
fila_atendimento.append('Paciente_4')
```

```python
# Removendo o Paciente 1, pois ele foi o primeiro a entrar na fila:
fila_atendimento.pop(0)

# Apresentaçao:
print(fila_atendimento)
```

```python
# Removendo o Paciente 2, como o Paciente 1 já saiu da fila o Paciente 2 está na posição 0:
fila_atendimento.pop(0)

# Apresentaçao:
print(fila_atendimento)
```
