# Step-1
- Use a lista de números: [12, 45, 67, 23, 89, 34, 22, 90, 56, 78].

- Calcule a média dos números usando a biblioteca numpy.

```python
# Creating a list of values:
numbers = [12, 45, 67, 23, 89, 34, 22, 90, 56, 78]
```

$$
\text{Média} = \frac{\sum_{i=1}^{n} x_i}{n}
$$

```python
# Calculating the average of the numbers from the equation:
mean_numbers = sum(numbers) / len(numbers)

# Average presentation:
print(f"A média dos valores {numbers} é {mean_numbers}")
```

```python
# Calculating the average of numbers using the library:
mean_numbers = np.mean(numbers)

# Average presentation:
print(f"A média dos valores {numbers} é {mean_numbers}")
```

# Step-2
- Identifique os números que são maiores que a média.

- Armazene os números maiores que a média em um DataFrame do pandas.

```python
'''
                                Alternativa 1 
'''
# Identifying numbers that are greater than the average using a filter and the lambda function (code reduction):
numbers_maior_mean = list(filter(lambda x: x>mean_numbers, numbers))

# Creating the Dataframe:
numeros_maiores_que_media = pd.DataFrame(numbers_maior_mean, columns=['numeros_maiores_média'])

# Presentation of results:
print(f'Números maiores que a média: {numbers_maior_mean}')

print(f"Dataframe com os números da lista 'numbers' maiores que a méda {mean_numbers}")
display(numeros_maiores_que_media)
```

```python
'''
                                Alternativa 2
'''
# Dataframe to store values ​​greater than the average:
df_numbers_maior_mean = pd.DataFrame(columns=['numeros_maiores_média'])

# Applying a loop to traverse the list of 'numbers' elements:
for number in numbers:
    if number > mean_numbers:
        df_numbers_maior_mean = df_numbers_maior_mean.append({'numeros_maiores_média': number}, ignore_index=True)
        
# Presentation of results
print(f"Dataframe com os números da lista 'numbers' maiores que a méda {mean_numbers}")
display(df_numbers_maior_mean)        

```

# Step-3
- Salve o DataFrame em um arquivo CSV chamado "numeros_maiores_que_media.csv"

```python
# Saving the DataFrame to a CSV file
df_numbers_maior_mean.to_csv("numeros_maiores_que_media.csv", index=False)
print('Arquivo salvo com sucesso :)')
```
