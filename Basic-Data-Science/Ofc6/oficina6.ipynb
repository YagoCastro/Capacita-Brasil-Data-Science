## 0.2.Dados_Colletion

```python
dados = {

    'Nome': ['Ana', 'Bruno', 'Carlos', 'Daniela', 'Eduardo'],

    'Idade': [28, 34, 29, None, 42],

    'Cidade': ['São Paulo', 'Rio de Janeiro', None, 'Curitiba', 'Belo Horizonte'],

    'Vendas': [150, 200, 300, 400, None]

}
```

# P1.:Crie um DataFrame com base nos dados fornecidos;

```python
df = pd.DataFrame(dados)
```

```python
df.head()
```

# P2.:Filtre os clientes que têm mais de 30 anos;

```python
#print(f"Os clientes que têm mais de 30 anos: \n\n {df[df['Idade'] > 30]}")
print(f"Os clientes que têm mais de 30 anos: ")
display(df[df['Idade'] > 30])
```

# P3.:Calcule a média de idade dos clientes;

```python
df['Idade'].mean()
```

```python
np.mean(df['Idade'])
```

```python
idade = df['Idade']
idade.mean()
```

```python
soma = df['Idade'].sum() 
quantidade = df['Idade'].count() 
media = soma / quantidade  

print(f"Média das idades: {media}")
```

```python
soma = 0
contador = 0

for idade in df['Idade']:
    if idade is not None:  
        soma += idade
        contador += 1

# Calculando a média
media = soma / contador if contador > 0 else 0

print(f"Média das idades: {media}")
```

# P4.:Substitua valores faltantes na coluna Cidade por ‘Desconhecido’;

```python
df.head()
```

```python
df['Cidade'] = df['Cidade'].fillna('Desconhecido')
```

```python
df.head()
```
