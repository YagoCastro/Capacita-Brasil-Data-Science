Como você poderia calcular a probabilidade de um cliente esperar mais de 10 minutos na fila, baseado nos dados históricos?

Poderia inicialmente ajustar a distribuição da variável para uma distribuição normal ou exponencial, e depois calcular a probabilidade 
𝑃(𝑋>10)P(X>10):

$$
P(X > 10) = 1 - P(X ≤ 10) = 1 - Φ\left(\frac{10 - \mu}{\sigma}\right)
$$
Onde Φ é a função de distribuição acumulada (CDF) da distribuição normal, 
μ é a média e σ é o desvio padrão.
- E, para calcular a probabilidade desconsiderando o ajuste dos dados::

$$
P(X > 10) = \frac{N observações onde X > 10}{Total de observações}
