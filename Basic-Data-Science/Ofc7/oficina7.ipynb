### O conceito de Banco de Dados

Um banco de dados: **consiste em um conjunto de dados que são necessários para uma empresa ou órgão público organizarem os serviços. Sob essa ótica, é possível considerar a planilha de consumidores de um pequeno comércio e o caderno de anotações de uma lanchonete exemplos de banco de dados.** 
Por outro lado, a transformação digital faz com que o conhecimento tenha um valor cada vez mais estratégico. Em razão disso, os bancos de dados precisam permitir o acesso às informações de forma organizada, simples, ágil e segura. [1]

Esse cenário cria um ambiente favorável para o uso do Sistema de Gerenciamento de Banco de Dados (SGBD), cujo nome, em inglês, é Data Base Management System (DBMS).Vale destacar que o SGBD é um sistema, responsável por permitir que informações sejam armazenadas, estruturadas, atualizadas, excluídas, inseridas e protegidas, de acordo com a demanda. [1]

# Questão 2: 
Pesquise as diferenças entre bancos de dados relacionais e NoSQL. Quais tipos de empresas se beneficiam mais de cada um?

## O que é SQL?

O SQL (Structured Query Language) é uma linguagem de programação utilizada para gerenciar bancos de dados relacionais. Bancos de dados relacionais são aqueles que armazenam dados em tabelas, onde cada tabela é composta por linhas e colunas. As tabelas possuem chaves primárias que identificam exclusivamente cada linha e as relações entre as tabelas são gerenciadas por chaves estrangeiras.[3]

A linguagem é extremamente poderosa e permite aos usuários realizar consultas complexas em bancos de dados relacionais. Além disso, é padrão, o que significa que é amplamente aceita e utilizada em todo o mundo.[3]

## O que é NoSQL?

NoSQL, por outro lado, é uma categoria de bancos de dados não relacionais que não utilizam a linguagem SQL. Eles são projetados para armazenar grandes volumes de dados que não se encaixam bem em um modelo de tabela tradicional.[3]

Os bancos de dados NoSQL são altamente escaláveis e são frequentemente utilizados em aplicações web e móveis que precisam lidar com o excesso de dados. Eles são projetados para serem distribuídos em múltiplas máquinas, o que os torna altamente disponíveis e tolerantes a falhas.[3]

## Diferenças entre SQL e NoSQL

A principal diferença entre SQL e NoSQL é a forma como os dados são armazenados. O SQL utiliza um modelo de tabela relacional, enquanto o NoSQL utiliza um modelo de documento, chave-valor ou gráfico.[3]

O modelo de tabela relacional do SQL é altamente estruturado e requer que os dados sejam organizados em tabelas com esquemas pré-definidos. O modelo de documento do NoSQL, por outro lado, é mais flexível e permite que os dados sejam armazenados em documentos que não precisam seguir um esquema rígido.[3]

Outra diferença importante é a escalabilidade. Os bancos de dados NoSQL são altamente escaláveis e podem ser distribuídos em vários servidores, o que permite que eles gerenciem grandes volumes de dados. Os bancos de dados SQL, por outro lado, não são tão escaláveis e podem ter restrições de tamanho devido à sua estrutura rígida.[3]

## Quando utilizar SQL e NoSQL?

A escolha entre SQL e NoSQL deve ser feita com base nas necessidades da sua aplicação. Se a sua aplicação precisa armazenar dados altamente estruturados em tabelas, o SQL é a melhor escolha. O SQL é ideal para aplicativos que precisam de transações consistentes e precisas, como sistemas bancários e de reservas.[3]

Por outro lado, se a sua aplicação precisa armazenar grandes volumes de dados não estruturados, como imagens, vídeos e documentos, o NoSQL é a melhor escolha. O NoSQL é ideal para aplicativos que precisam de alta disponibilidade e escalabilidade, como redes sociais e aplicativos de comércio eletrônico.[3]

## Exemplos de bancos de dados relacionais (SQL)

Há vários bancos de dados relacionais populares disponíveis no mercado. Aqui estão alguns exemplos de bancos de dados relacionais:

**MySQL :** MySQL é um sistema de gerenciamento de banco de dados relacional de código aberto amplamente usado em aplicativos da web. É conhecido por sua velocidade, escalabilidade e facilidade de uso. [4]

**Oracle Database :** O Oracle Database é um sistema de gerenciamento de banco de dados relacional proprietário comumente usado em aplicativos de nível empresarial. Ele fornece forte suporte para conformidade com ACID, alta disponibilidade e escalabilidade. [4]

**Microsoft SQL Server :** O Microsoft SQL Server é um sistema de gerenciamento de banco de dados relacional comumente usado em ambientes baseados em Windows. Ele fornece forte suporte para aplicativos de nível empresarial, incluindo business intelligence e data warehousing.[4]

**PostgreSQL :** O PostgreSQL é um sistema de gerenciamento de banco de dados relacional de código aberto conhecido por sua robustez, flexibilidade e suporte a recursos avançados, como pesquisa de texto completo e dados geoespaciais. [4]

**SQLite :** SQLite é um sistema de gerenciamento de banco de dados relacional leve, baseado em arquivo, amplamente usado em aplicativos móveis e de desktop. É conhecido por sua simplicidade, confiabilidade e portabilidade. [4]

## Exemplos de bancos de dados não relacionais (NoSQL)

Há vários bancos de dados não relacionais populares disponíveis no mercado. Aqui estão alguns exemplos:

**Couchbase :** O Couchbase é um banco de dados distribuído que suporta modelos de dados de chave-valor e documento. Ele é projetado para alta escalabilidade, desempenho e disponibilidade e suporta recursos como auto-sharding, cache na memória e pesquisa de texto completo. O Couchbase é bem adequado para lidar com grandes conjuntos de dados e alto rendimento de gravação, tornando-o popular para aplicativos de comércio eletrônico, jogos e mídia social.[4]

**MongoDB :** Um banco de dados orientado a documentos que armazena dados em documentos do tipo JSON.[4]

**Apache Cassandra :** Um banco de dados distribuído que armazena dados em um formato de família de colunas.[4]

**Redis :** um armazenamento de chave-valor que pode ser usado como banco de dados, cache e corretor de mensagens.[4]

**Amazon DynamoDB :** um serviço de banco de dados NoSQL gerenciado fornecido pela Amazon Web Services (AWS).[4]

**Neo4j :** Um banco de dados gráfico que armazena dados em nós e arestas.[4]

## Funcionamento do Sistema

- A tabela Produtos mantém um catálogo de todos os itens disponíveis para venda.

- A tabela Clientes registra os dados demográficos e financeiros dos clientes para análise e personalização de marketing.

- A tabela Vendas registra cada transação realizada, incluindo os detalhes sobre os produtos comprados, os clientes que realizaram as compras e os valores envolvidos.

Esse é um banco de dados com um diagrama de entidades e relacionamentos, utilizando variáveis simples, ou seja, entradas iniciais em um banco de dados pequeno. ## Relacionamentos

- Produtos ↔ Vendas: A tabela Vendas usa o campo IDProduto como chave estrangeira para vincular cada venda a um produto específico da tabela Produtos.

- Clientes ↔ Vendas: A tabela Vendas também usa o campo IDConsumidor como chave estrangeira para vincular cada venda a um cliente específico da tabela Clientes.
