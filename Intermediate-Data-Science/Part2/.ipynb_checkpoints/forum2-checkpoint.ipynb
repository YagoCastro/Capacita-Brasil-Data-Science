Essa é um técnica bastante explorada hoje pelos mercados para aumentar a venda de determinados produtos verificando as transações de vendas dos clientes e quais produtos estão saindo na mesma sacola de compras. Isso também permitindo a criação de estratégias de marketing mais eficazes, como promoções mais especificas ou um atendimento mais personalizado
