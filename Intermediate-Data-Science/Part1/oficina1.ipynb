Você foi recentemente contratado como analista de dados em uma empresa de tecnologia que desenvolve soluções para a gestão de operações em empresas de diversos setores. Parte do seu trabalho envolve a análise de relatórios sobre incidentes que ocorreram em diferentes instalações da empresa. 

Esses relatórios são importantes, pois ajudam os gestores a identificar padrões, entender as causas dos incidentes e tomar decisões para prevenir problemas futuros.

O relatório atual apresenta gráficos que incluem dados, como status dos incidentes, número de incidentes por instalação, categorias de incidentes e causas contribuintes. Ao examinar os gráficos e relatórios fornecidos, você se deparou com estas informações:

# Item B
Como esses problemas podem impactar a interpretação dos dados pelos tomadores de decisão?

<font color = "blue"> **Resposta:**</font>

Deixar esses problemas inseridos no dashboard de apresentação dos resultados pode inibir ou dificultar a visualização de algumas possíveis ações que poderiam fazer parte para solucionar o problema. 

<a id="itemc"></a>
# Item C
Para mostrar a comparação de desempenho entre diferentes tipos de incidentes, qual tipo de gráfico você recomendaria e por quê?

<font color = "blue"> **Resposta:**</font>

Poderia ser o gráfico de barras agrupadas, ele vai permitir comparar facilmente a frequência ou impacto de diferentes tipos de incidentes em uma escala clara e organizada. Poderia também incluir múltiplas variáveis, como severidade ou status (aberto/fechado), agrupadas para cada tipo de incidente.

<a id="itemd"></a>
# Item D
Se você precisasse exibir a evolução dos incidentes ao longo do tempo e o status dela para diferentes grupos demográficos, qual seria o tipo de gráfico mais apropriado?

<font color = "blue"> **Resposta:**</font>

Poderia ser um Gráfico de Linhas Empilhadas ou Linhas com Múltiplas Séries. Dependendo da quantidade de classes poderia ser um único gráfico ou multiplos gráficos com a mesma escala.
