<a id="Gerando"></a>
## Gerando os dados

```python
# Declarando as listas
TurmaT1_Mat     = []
TurmaT1_Port    = []
TurmaT1_Science = []

TurmaT2_Mat     = []
TurmaT2_Port    = []
TurmaT2_Science = []

TurmaT3_Mat     = []
TurmaT3_Port    = []
TurmaT3_Science = []

# Definindo uma semente de aleatoriedade fixa, para que as análises nos
# próximos passos tenham 
random.seed(10)

# Declarnado a quantidade de valores em cada lista:
n = 50

# Relizando um laço for para o preenchimento das listas com os valores:
for i in range(n):
    TurmaT1_Mat.append(random.choice(list(range(11)) ))
    TurmaT1_Port.append(random.choice(list(range(11)) ))
    TurmaT1_Science.append(random.choice(list(range(11)) ))
    
    TurmaT2_Mat.append(random.choice(list(range(11)) ))
    TurmaT2_Port.append(random.choice(list(range(11)) ))
    TurmaT2_Science.append(random.choice(list(range(11)) ))
    
    TurmaT3_Mat.append(random.choice(list(range(11)) ))
    TurmaT3_Port.append(random.choice(list(range(11)) ))
    TurmaT3_Science.append(random.choice(list(range(11)) ))

    
print(f'Notas TurmaT1_Mat:\n\n{TurmaT1_Mat}\n')
print(f'Notas TurmaT1_Port:\n\n{TurmaT1_Port}\n')
print(f'Notas TurmaT1_Science:\n\n{TurmaT1_Science}\n')

print(f'Notas TurmaT2_Mat:\n\n{TurmaT2_Mat}\n')
print(f'Notas TurmaT2_Port:\n\n{TurmaT2_Port}\n')
print(f'Notas TurmaT2_Science:\n\n{TurmaT2_Science}\n')

print(f'Notas TurmaT1_Mat:\n\n{TurmaT1_Mat}\n')
print(f'Notas TurmaT2_Port:\n\n{TurmaT2_Port}\n')
print(f'Notas TurmaT3_Science:\n\n{TurmaT3_Science}')
```

<a id="Dataframe"></a>
## Transformando em um único Dataframe

```python
# Transformando as listas em Dataframes:
df_TurmaT1_Mat    = pd.DataFrame(TurmaT1_Mat)
df_TurmaT1_Port   = pd.DataFrame(TurmaT1_Port)
df_TurmaT1_Science= pd.DataFrame(TurmaT1_Science)

df_TurmaT2_Mat    = pd.DataFrame(TurmaT2_Mat)
df_TurmaT2_Port   = pd.DataFrame(TurmaT2_Port)
df_TurmaT2_Science= pd.DataFrame(TurmaT2_Science)

df_TurmaT3_Mat    = pd.DataFrame(TurmaT3_Mat)
df_TurmaT3_Port   = pd.DataFrame(TurmaT3_Port)
df_TurmaT3_Science= pd.DataFrame(TurmaT3_Science)
```

```python
# Juntando os Datasets:
df_Turmas = pd.concat([
    df_TurmaT1_Mat,    
    df_TurmaT1_Port,   
    df_TurmaT1_Science,
    df_TurmaT2_Mat,   
    df_TurmaT2_Port,  
    df_TurmaT2_Science,
    df_TurmaT3_Mat,    
    df_TurmaT3_Port,   
    df_TurmaT3_Science
], axis = 1)
```

```python
df_Turmas
```

```python
df_Turmas.columns = [
    'TurmaT1_Mat',    
    'TurmaT1_Port',   
    'TurmaT1_Science',
    'TurmaT2_Mat',   
    'TurmaT2_Port',   
    'TurmaT2_Science',
    'TurmaT3_Mat',    
    'TurmaT3_Port',  
    'TurmaT3_Science']
```

```python
df_Turmas
```

## Gerando os gráficos

```python
# Configuração do estilo do gráfico
plt.style.use('seaborn')
plt.figure(figsize=(10, 6))

# Lista de colunas e cores
turmas = ['TurmaT1_Mat', 'TurmaT2_Mat', 'TurmaT3_Mat']
cores = ['blue', 'green', 'orange']

# Adicionar as distribuições ao mesmo gráfico
for turma, cor in zip(turmas, cores):
    sns.kdeplot(data=df_Turmas[turma], label=turma, color=cor, fill=True, alpha=0.3)

# Configurar título e legendas
plt.title('Distribuição das Notas por Turma', fontsize=14)
plt.xlabel('Notas', fontsize=12)
plt.ylabel('Densidade', fontsize=12)
plt.legend(title='Turmas', fontsize=10)
plt.grid(True)

# Exibir o gráfico
plt.show()
```

```python

# Calcular as médias das turmas
turmas = ['TurmaT1_Mat', 'TurmaT2_Mat', 'TurmaT3_Mat']
medias = [df_Turmas[turma].mean() for turma in turmas]

# Configurar o gráfico
plt.figure(figsize=(8, 6))
plt.bar(turmas, medias, color=['blue', 'green', 'orange'], alpha=0.8)

# Adicionar rótulos e título
plt.title('Comparação de Médias das Notas por Turma', fontsize=14)
plt.xlabel('Turmas', fontsize=12)
plt.ylabel('Média das Notas', fontsize=12)

# Adicionar valores no topo das barras
for i, media in enumerate(medias):
    plt.text(i, media + 0.1, f'{media:.2f}', ha='center', fontsize=10)

# Exibir o gráfico
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Configuração do estilo do gráfico
plt.style.use('seaborn')
plt.figure(figsize=(10, 6))

# Criar o boxplot
sns.boxplot(data=df_Turmas[['TurmaT1_Mat', 'TurmaT2_Mat', 'TurmaT3_Mat']], palette='Set2')

# Configurar título e eixos
plt.title('Distribuição de Notas por Turma (Boxplot)', fontsize=14)
plt.xlabel('Turmas', fontsize=12)
plt.ylabel('Notas', fontsize=12)

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
# Configuração do estilo do gráfico
plt.style.use('seaborn')
plt.figure(figsize=(10, 6))

# Traçar as linhas para cada turma
for turma, cor in zip(['TurmaT1_Mat', 'TurmaT2_Mat', 'TurmaT3_Mat'], ['blue', 'green', 'orange']):
    plt.plot(df_Turmas.index, df_Turmas[turma], label=turma, color=cor, marker='o')

# Configurar título e eixos
plt.title('Comparação de Notas por Turma', fontsize=14)
plt.xlabel('Índice', fontsize=12)
plt.ylabel('Notas', fontsize=12)
plt.legend(title='Turmas', fontsize=10)

# Exibir o gráfico
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Gráfico de Dispersão (Scatter Plot)
plt.figure(figsize=(12, 6))
plt.scatter(range(len(df_Turmas)), df_Turmas["TurmaT1_Mat"], label="TurmaT1_Mat", color="blue", alpha=0.7)
plt.scatter(range(len(df_Turmas)), df_Turmas["TurmaT2_Mat"], label="TurmaT2_Mat", color="green", alpha=0.7)
plt.scatter(range(len(df_Turmas)), df_Turmas["TurmaT3_Mat"], label="TurmaT3_Mat", color="purple", alpha=0.7)
plt.title("Gráfico de Dispersão das Notas por Turma - Matemática")
plt.xlabel("Estudantes")
plt.ylabel("Notas")
plt.xticks(range(len(df_Turmas)), [f"Aluno {i+1}" for i in range(len(df_Turmas))], rotation=90)
plt.legend()
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
