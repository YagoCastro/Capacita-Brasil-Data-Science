# Questão 1
Como você analisa a aplicação da análise de sentimentos em situações do dia a dia? Pode pensar em exemplos práticos em que essa técnica seria útil?



Essa aplicação seria muito útil no atendimento ao cliente, melhorando cada vez mais a experiência do consumidor com o serviço ou a loja. No recrutamento e recursos humanos, o uso de PLN para analisar sentimentos em avaliações de funcionários sobre a cultura da empresa pode ajudar na gestão organizacional.

Um caso bem interessante é na área de saúde mental e bem-estar, onde é possível analisar sentimentos e identificar sinais de estresse ou depressão em textos escritos, como postagens em redes sociais ou interações com chatbots de suporte emocional.


# Questão 2
Chatbots estão cada vez mais presentes em serviços que usamos. Você se lembra de alguma interação marcante com um chatbot? Como essa experiência lhe marcou?

Na verdade, constantemente venho usando inteligências artificiais e chatbots para complementar meu aprendizado, trazendo novas perspectivas ou conteúdos sobre o assunto mencionado. Acredito que algo marcante foi na construção de imagens para uma apresentação no Canva, onde todas as imagens da apresentação foram geradas com uma temática específica e bem característica. Isso gerou até perguntas sobre onde encontramos as imagens para o trabalho, e essa parte foi bastante interessante, pois conseguimos manter um mesmo design durante todo o trabalho, algo que, ao pesquisar imagens, é bastante difícil. Outro exemplo foi com as novas funcionalidades do ChatGPT, como a leitura de PDFs e até a geração de imagens.

# Questão 3
As palavras têm significados complexos que dependem de contexto. Em sua opinião, quais desafios surgem ao tentar ensinar uma máquina a entender nuances e contextos em textos?

Acredito que as próprias variações da língua, especialmente no português, representam um desafio, pois cada região possui gírias e expressões características. Além disso, há elementos como o sarcasmo e a ironia, que dificultam a interpretação por parte das máquinas. Outros desafios incluem a ambiguidade semântica, a dependência do contexto e as mudanças rápidas de contexto.
